```python
trainSplit = int(len(X) * 0.6)
devSplit = int(len(X) * 0.8)
x_train = X.iloc[: trainSplit, :]
y_train  = Y.iloc[: trainSplit]
x_dev = X.iloc[trainSplit : devSplit, :]
y_dev = Y.iloc[trainSplit : devSplit]
x_test = X.iloc[devSplit: , :]
y_test = Y.iloc[devSplit:]
```

```python
model = nn.Sequential(nn.Linear(x_train.shape[1], 30),
                      nn.ReLU(), 
                      nn.Linear(30, 20),
                      nn.ReLU(),
                      nn.Linear(20, 10),
                      nn.ReLU(),
                      nn.Linear(10,1))
```

```python
loss_funct = torch.nn.MSELoss()
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr = 0.01)
```
